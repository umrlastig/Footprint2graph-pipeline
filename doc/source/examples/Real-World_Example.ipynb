```python
import os
import sys

import matplotlib.pyplot as plt

# Import de tracklib
module_path = os.path.abspath(os.path.join('../../../../tracklib'))
if module_path not in sys.path:
    sys.path.append(module_path)

# Import de OFNP
module_path = os.path.abspath(os.path.join('../../..'))
if module_path not in sys.path:
    sys.path.append(module_path)

import tracklib as tkl
```

```python
chemin = os.path.abspath(os.path.join('../../../data/'))

# 2023-07-15T15:18:21.000Z
tkl.ObsTime.setReadFormat("4Y-2M-2DT2h:2m:2sZ")

param = tkl.TrackFormat({'ext': 'GPX', 'srid': 'ENU', 'type': 'trk'})
tracks = tkl.TrackReader.readFromFile(chemin, param)

print (tracks.size())
tracks.plot()
```
